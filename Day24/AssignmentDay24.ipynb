# Project 1:  

 **Build Decision Tree(DV-"Survived",IDV-"Age,Gender and Fare") and Prediction**

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
```

```python
titanic_train = pd.read_csv(r'C:\Users\karth\Assignments\Day24\Data_decision_tree\train.csv')
titanic_train.head()
```

```python
titanic_train['Age'].mean()
```

```python
new_age_var = np.where(titanic_train['Age'].isnull(),32,titanic_train['Age'])

titanic_train['Age'] = new_age_var
```

```python
titanic_train['Age']
```

```python
titanic_train.info()
```

### Label Encoding

```python
label_encoder = preprocessing.LabelEncoder()
encoded_sex = label_encoder.fit_transform(titanic_train['Sex'])
```

### Checking null and na values

```python
print(titanic_train['Age'].isnull().sum())
print(titanic_train['Age'].isna().sum())

print(titanic_train['Sex'].isnull().sum())
print(titanic_train['Sex'].isna().sum())

print(titanic_train['Fare'].isnull().sum())
print(titanic_train['Fare'].isna().sum())
```

### Initialising Classifier

```python
tree_model = tree.DecisionTreeClassifier()
```

**Note: Considering only Age, Gender, Fare as Independent variables and Survived as Dependednt variables**

```python
predictors = pd.DataFrame([titanic_train['Age'],encoded_sex,titanic_train['Fare']]).T

predictors = predictors.rename({'Unnamed 0': 'Gender'}, axis=1)
predictors
```

```python
tree_model = tree.DecisionTreeClassifier(max_depth=6)

tree_model.fit(X=predictors,y=titanic_train['Survived'])
```

### Classfication tree result and score:

```python
with open("Dtree.dot",'w') as f:
    f = tree.export_graphviz(tree_model,feature_names=['Age','Gender','Fare'],out_file=f);
```

Note: Converted the Dot file code in to Classification to visualise in WebGraphviz.com and saved as Image.

```python
from IPython.display import display, Image
display(Image(filename='Dtree.png'))
```

```python
tree_model.score(X=predictors,y=titanic_train['Survived'])
```

## Testing with Test Data:

```python
titanic_test = pd.read_csv(r'C:\Users\karth\Assignments\Day24\Data_decision_tree\test.csv')
```

```python
new_age_var = np.where(titanic_test['Age'].isnull(),32,titanic_test['Age'])

titanic_test['Age'] = new_age_var
```

```python
label_encoder = preprocessing.LabelEncoder()
encoded_sex = label_encoder.fit_transform(titanic_test['Sex'])
```

```python
test_features= pd.DataFrame([titanic_test['Age'],encoded_sex,titanic_test['Fare']]).T

test_features = test_features.rename({'Unnamed 0': 'Gender'}, axis=1)
test_features
```

```python
test_preds = tree_model.predict(X=test_features)
```

```python
predicted_output = pd.DataFrame({"PassengerID":titanic_test["PassengerId"],"Survived":test_preds})
```
